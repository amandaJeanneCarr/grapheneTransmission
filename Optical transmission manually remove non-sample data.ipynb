```python
# Code to manually separate non-graphene sample data post-transfer from graphene sample data to generate detected and
# incident power background information

%matplotlib inline
%run get_heatmap.ipynb

#Import necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

# Load det/inc power ratio data for graphene/polymer stack. Plot
samp='p_EVA_toluene'; letter='1'; date='2019_07_24'; pixel=0.5;
path='C:\\Users\\acarr.BNL\\Documents\\Data\\Optical Transmission\\'; 
stackPower=pd.read_csv(path+date+'\\Gr_'+samp+'_'+letter+'_fixed.csv');
plt.title(samp+'_'+letter);
plt.show(sb.heatmap(stackPower));

# Enter indices of non-graphene sample areas. These can include mirror data
rowS1=0; rowE1=40; colmS1=0; colmE1=40;
rowS2=0; rowE2=40; colmS2=0; colmE2=40;
rowS3=0; rowE3=40; colmS3=0; colmE3=40;
rowS4=0; rowE4=40; colmS4=0; colmE4=40;
# Use indices to obtain non-graphene sample data from graphene/polymer stack power ratio data. Dump into an array.
# Eliminate mirror data values by excluding power ratios over 0.3
dum1=stackPower.iloc[rowS1:rowE1,colmS1:colmE1].values.flatten();
dum2=stackPower.iloc[rowS2:rowE2,colmS2:colmE2].values.flatten(); 
dum3=stackPower.iloc[rowS3:rowE3,colmS3:colmE3].values.flatten(); 
dum4=stackPower.iloc[rowS4:rowE4,colmS4:colmE4].values.flatten(); 
powerArray=np.hstack((dum1,dum2,dum3,dum4)); 
# Repeat process for incident power data using index from power ratio data to eliminate mirror data. Save to file
stackInc=pd.read_csv(path+date+'\\Gr_'+samp+'_'+letter+'_inc_fixed.csv'); 
dum1=stackInc.iloc[rowS1:rowE1,colmS1:colmE1].values.flatten(); 
dum2=stackInc.iloc[rowS2:rowE2,colmS2:colmE2].values.flatten(); 
dum3=stackInc.iloc[rowS3:rowE3,colmS3:colmE3].values.flatten(); 
dum4=stackInc.iloc[rowS4:rowE4,colmS4:colmE4].values.flatten(); 
incArray=np.hstack((dum1,dum2,dum3,dum4)); A=incArray[(powerArray<0.3)&(powerArray>0)]; A=pd.DataFrame(A); 
A.to_csv(path+samp+letter+'_avgInc.csv',sep=',',header=False,index=False)

stackDet=pd.read_csv(path+date+'\\Gr_'+samp+'_'+letter+'_det_fixed.csv');
dum1=stackDet.iloc[rowS1:rowE1,colmS1:colmE1].values.flatten(); 
dum2=stackDet.iloc[rowS2:rowE2,colmS2:colmE2].values.flatten(); 
dum3=stackDet.iloc[rowS3:rowE3,colmS3:colmE3].values.flatten(); 
dum4=stackDet.iloc[rowS4:rowE4,colmS4:colmE4].values.flatten(); 
detArray=np.hstack((dum1,dum2,dum3,dum4)); A=detArray[(powerArray<0.3)&(powerArray>0)]; A=pd.DataFrame(A); 
A.to_csv(path+samp+letter+'_avgDet.csv',sep=',',header=False,index=False)
```
